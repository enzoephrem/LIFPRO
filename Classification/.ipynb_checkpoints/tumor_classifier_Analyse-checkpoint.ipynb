# Brain Tumor Classification Analyse

## Analyse Resluts and Graphics

```python
from packaging import version

import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from scipy import stats
import tensorboard as tb
```

```python
experiment_id = "qapkhPFBRsa2i7O4vYLL6Q"
experiment = tb.data.experimental.ExperimentFromDev(experiment_id)
df = experiment.get_scalars()
df = df[df.tag != 'evaluation_accuracy_vs_iterations']
df = df[df.tag != 'evaluation_loss_vs_iterations']
df.reset_index(inplace=True)
df
```

```python
dfw = df.pivot(index=['run','step'], columns=['tag'])['value']
dfw.reset_index(inplace=True)
dfw
```

```python
csv_path = 'tableau_logs.csv'
dfw.to_csv(csv_path, index=False)
```

```python
# Filter the DataFrame to only validation data, which is what the subsequent
# analyses and visualization will be focused on.
dfw_validation =  dfw[dfw.run.str.contains("validation")]
# Get the optimizer value for each row of the validation DataFrame.
optimizer_validation = dfw_validation.run.apply(lambda run: run.split("_")[1] )

plt.figure(figsize=(16, 6))
plt.subplot(1, 2, 1)
sns.lineplot(data=dfw_validation, x="step", y="epoch_accuracy",
             hue=optimizer_validation).set_title("Accuracy of validation values according to the pre-model")
plt.subplot(1, 2, 2)
sns.lineplot(data=dfw_validation, x="step", y="epoch_loss",
             hue=optimizer_validation).set_title("Loss of of validation values according to the pre-model")
plt.savefig('premodel_1.png', transparent=False)
```

```python
plt.figure(figsize=(16, 6))
plt.subplot(1, 2, 1)
adam_min_val_loss = dfw_validation.loc[optimizer_validation=="EfficientNetB0", :].groupby(
    "run", as_index=False).agg({"epoch_accuracy": "min"})
rmsprop_min_val_loss = dfw_validation.loc[optimizer_validation=="MobileNetV2", :].groupby(
    "run", as_index=False).agg({"epoch_accuracy": "min"})
sgd_min_val_loss = dfw_validation.loc[optimizer_validation=="ResNet", :].groupby(
    "run", as_index=False).agg({"epoch_accuracy": "min"})
vgg_min_val_loss = dfw_validation.loc[optimizer_validation=="VGG16", :].groupby(
    "run", as_index=False).agg({"epoch_accuracy": "min"})
min_val_loss = pd.concat([adam_min_val_loss, rmsprop_min_val_loss, sgd_min_val_loss,vgg_min_val_loss])
sns.boxplot(data=min_val_loss, y="epoch_accuracy",
            x=min_val_loss.run.apply(lambda run: run.split("_")[1])).set_title("Accuracy of validation values according to the pre-model")

plt.subplot(1, 2, 2)
adam_min_val_loss = dfw_validation.loc[optimizer_validation=="EfficientNetB0", :].groupby(
    "run", as_index=False).agg({"epoch_loss": "min"})
rmsprop_min_val_loss = dfw_validation.loc[optimizer_validation=="MobileNetV2", :].groupby(
    "run", as_index=False).agg({"epoch_loss": "min"})
sgd_min_val_loss = dfw_validation.loc[optimizer_validation=="ResNet", :].groupby(
    "run", as_index=False).agg({"epoch_loss": "min"})
vgg_min_val_loss = dfw_validation.loc[optimizer_validation=="VGG16", :].groupby(
    "run", as_index=False).agg({"epoch_loss": "min"})
min_val_loss = pd.concat([adam_min_val_loss, rmsprop_min_val_loss, sgd_min_val_loss, vgg_min_val_loss])
sns.boxplot(data=min_val_loss, y="epoch_loss",
            x=min_val_loss.run.apply(lambda run: run.split("_")[1])).set_title("Loss of of validation values according to the pre-model")
plt.savefig('premodel_2.png', transparent=False)
```

```python
# Filter the DataFrame to only validation data, which is what the subsequent
# analyses and visualization will be focused on.
dfw_mobile =  dfw_validation[dfw_validation.run.str.contains("VGG16")]\

# Get the optimizer value for each row of the validation DataFrame.
optimizer_mobile = dfw_mobile.run.apply(lambda run: run.split("_")[2] )

plt.figure(figsize=(16, 6))
plt.subplot(1, 2, 1)
sns.lineplot(data=dfw_mobile, x="step", y="epoch_accuracy",
             hue=optimizer_mobile).set_title("Accuracy of validation values and pre-model VGG16" + '\n'+ "according to the optimizer function")
plt.subplot(1, 2, 2)
sns.lineplot(data=dfw_mobile, x="step", y="epoch_loss",
             hue=optimizer_mobile).set_title("Loss of validation values and pre-model VGG16" + '\n'+ "according to the optimizer function")
plt.savefig('optimizer_1.png', transparent=False)
```

```python
plt.figure(figsize=(16, 6))
plt.subplot(1, 2, 1)
adam_min_val_loss = dfw_mobile.loc[optimizer_mobile=="SGD", :].groupby(
    "run", as_index=False).agg({"epoch_accuracy": "min"})
rmsprop_min_val_loss = dfw_mobile.loc[optimizer_mobile=="Adam", :].groupby(
    "run", as_index=False).agg({"epoch_accuracy": "min"})
sgd_min_val_loss = dfw_mobile.loc[optimizer_mobile=="NAdam", :].groupby(
    "run", as_index=False).agg({"epoch_accuracy": "min"})
min_val_loss = pd.concat([adam_min_val_loss, rmsprop_min_val_loss, sgd_min_val_loss])
sns.boxplot(data=min_val_loss, y="epoch_accuracy",
            x=min_val_loss.run.apply(lambda run: run.split("_")[2])).set_title("Accuracy of validation values and pre-model VGG16" + '\n'+ "according to the optimizer function")

plt.subplot(1, 2, 2)
adam_min_val_loss = dfw_mobile.loc[optimizer_mobile=="SGD", :].groupby(
    "run", as_index=False).agg({"epoch_loss": "min"})
rmsprop_min_val_loss = dfw_mobile.loc[optimizer_mobile=="Adam", :].groupby(
    "run", as_index=False).agg({"epoch_loss": "min"})
sgd_min_val_loss = dfw_mobile.loc[optimizer_mobile=="NAdam", :].groupby(
    "run", as_index=False).agg({"epoch_loss": "min"})
min_val_loss = pd.concat([adam_min_val_loss, rmsprop_min_val_loss, sgd_min_val_loss])
sns.boxplot(data=min_val_loss, y="epoch_loss",
            x=min_val_loss.run.apply(lambda run: run.split("_")[2])).set_title("Loss of validation values and pre-model VGG16" + '\n'+ "according to the optimizer function")

plt.savefig('optimizer_2.png', transparent=False)
```

```python
# Filter the DataFrame to only validation data, which is what the subsequent
# analyses and visualization will be focused on.
dfw_mobile =  dfw_validation[dfw_validation.run.str.contains("VGG16")]\

# Get the optimizer value for each row of the validation DataFrame.
loss_mobile = dfw_mobile.run.apply(lambda run: run.split("_")[4] )

loss_mobile = loss_mobile.str.replace('divergence', 'kl_divergence')
loss_mobile = loss_mobile.str.replace('hinge', 'categorical_hinge')
loss_mobile = loss_mobile.str.replace('crossentropy', 'categorical_crossentropy')

plt.figure(figsize=(16, 6))
plt.subplot(1, 2, 1)
sns.lineplot(data=dfw_mobile, x="step", y="epoch_accuracy",
             hue=loss_mobile).set_title(r"Accuracy of validation values and pre-model VGG16" + '\n'+ "according to the loss function")
plt.subplot(1, 2, 2)
sns.lineplot(data=dfw_mobile, x="step", y="epoch_loss",
             hue=loss_mobile).set_title("Loss of validation values and pre-model VGG16" + '\n'+ "according to the loss function")
plt.savefig('loss_1.png', transparent=False)
```

```python
plt.figure(figsize=(16, 6))
plt.subplot(1, 2, 1)
adam_min_val_loss = dfw_mobile.loc[loss_mobile=="kl_divergence", :].groupby(
    "run", as_index=False).agg({"epoch_accuracy": "min"})
rmsprop_min_val_loss = dfw_mobile.loc[loss_mobile=="categorical_hinge", :].groupby(
    "run", as_index=False).agg({"epoch_accuracy": "min"})
sgd_min_val_loss = dfw_mobile.loc[loss_mobile=="categorical_crossentropy", :].groupby(
    "run", as_index=False).agg({"epoch_accuracy": "min"})
min_val_loss = pd.concat([adam_min_val_loss, rmsprop_min_val_loss, sgd_min_val_loss])
sns.boxplot(data=min_val_loss, y="epoch_accuracy",
            x=min_val_loss.run.apply(lambda run: run.split("_")[4])).set_title("Accuracy of validation values and pre-model VGG16" + '\n'+ "according to the loss function")

plt.subplot(1, 2, 2)
adam_min_val_loss = dfw_mobile.loc[loss_mobile=="kl_divergence", :].groupby(
    "run", as_index=False).agg({"epoch_loss": "min"})
rmsprop_min_val_loss = dfw_mobile.loc[loss_mobile=="categorical_hinge", :].groupby(
    "run", as_index=False).agg({"epoch_loss": "min"})
sgd_min_val_loss = dfw_mobile.loc[loss_mobile=="categorical_crossentropy", :].groupby(
    "run", as_index=False).agg({"epoch_loss": "min"})
min_val_loss = pd.concat([adam_min_val_loss, rmsprop_min_val_loss, sgd_min_val_loss])
sns.boxplot(data=min_val_loss, y="epoch_loss",
            x=min_val_loss.run.apply(lambda run: run.split("_")[4])).set_title("Loss of validation values and pre-model VGG16" + '\n'+ "according to the loss function")
plt.savefig('loss_2.png', transparent=False)
```
